Part 3: Data Preprocessing & Model Building

Encode Categorical Variables

```python
df_model = df.copy()
```

```python
df_model['salary'] = (
    df_model['salary']
    .astype('category')
    .cat.set_categories(['low', 'medium', 'high'])
    .cat.codes
)
df_model = pd.get_dummies(df_model, columns=['department'], drop_first=True)
```

Verify the Dataset

```python
df_model.head()
```

Separate Features and Target

```python
X = df_model.drop('left', axis=1)
y = df_model['left']
```

Split the Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)
```

Check Dataset Shapes

```python
print("Training Features:", X_train.shape)
print("Testing Features :", X_test.shape)
print("Training Labels  :", y_train.shape)
print("Testing Labels   :", y_test.shape)
```

Train a Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    random_state=42,
    n_estimators=200
)

rf.fit(X_train, y_train)
```

Make Predictions

```python
y_pred = rf.predict(X_test)
```

valuate the Model
Accuracy

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

classification report

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

confusion metrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test)

plt.title("Confusion Matrix")
plt.show()
```

interpretation

The confusion matrix shows that the Random Forest model correctly classified most employees who stayed and those who left the company. There were very few false positives and false negatives, indicating excellent predictive performance.

feature importance

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf.feature_importances_
})

feature_importance = feature_importance.sort_values(
    by='Importance',
    ascending=False
)

feature_importance.head(10)
```

plot the top 10 features

```python
plt.figure(figsize=(8,6))

sns.barplot(
    data=feature_importance.head(10),
    x='Importance',
    y='Feature'
)

plt.title("Top 10 Important Features")
plt.show()
```

interpretation

Employee satisfaction is the strongest predictor of attrition, followed by the number of projects, tenure, average monthly working hours, and last evaluation score. These variables have the greatest influence on predicting whether an employee will leave the company.
